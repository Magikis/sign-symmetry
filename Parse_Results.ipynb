```python
%pylab inline
with open('/home/kku/Downloads/log_bckprop') as f:
    data = [x.split() for x in f]
```

```python
filtered = [
    x for x in data
    if 'Test:' in x and '[30/40]' in x
]


history = {
    'train_losses': [
        float(x[x.index('Prec@1') + 2][1:-1])
        for x in filtered
    ],
    'val_errs': [
        float(x[x.index('Prec@1') + 2][1:-1])
        for x in filtered
    ],
}
```

```python
def plot_history(history):
    figsize(16, 4)
    subplot(1,2,1)
    train_loss = np.array(history['train_losses'])
    semilogy(np.arange(train_loss.shape[0]), train_loss, label='batch train loss')
    legend()
        
    subplot(1,2,2)

    val_errs = np.array(history['val_errs'])[30:]
    plot(np.arange(val_errs.size), val_errs, label='validation error rate', color='r')
#     ylim(0,20)
    legend()
    print(max(val_errs))
```

```python
plot_history(history)
```
